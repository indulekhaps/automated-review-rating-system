<h1>Matplotlib Practice</h1>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

<h3>Line plot</h3>

```python
plt.plot(x, y)
plt.title("Line Plot")
plt.show()
```

<h3>Scatter plot</h3>

```python
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

<h3>Bar chart</h3>

```python
categories = ['A', 'B', 'C']
values = [10, 20, 15]
plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()
```

<h3>Horizontal bar chart</h3>

```python
plt.barh(categories, values)
plt.title("Horizontal Bar Chart")
plt.show()
```

<h3>Histogram</h3>

```python
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.title("Histogram")
plt.show()
```

<h3>Box plot</h3>

```python
plt.boxplot(data)
plt.title("Box Plot")
plt.show()
```

<h3>Pie chart</h3>

```python
plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

<h3>Multiple lines on one plot</h3>

```python
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.legend()
plt.title("Multiple Lines")
plt.show()
```

<h3>Subplots</h3>

```python
fig, axs = plt.subplots(2, 1)
axs[0].plot(x, np.sin(x))
axs[1].plot(x, np.cos(x))
plt.suptitle("Subplots Example")
plt.show()
```

<h3>Changing color and style</h3>

```python
plt.plot(x, y, color='pink', linestyle='--', marker='o')
plt.title("Styled Line")
plt.show()
```

<h3>Adding labels and title</h3>

```python
plt.plot(x, y,color='violet')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Labeled Plot")
plt.show()
```

<h3>Fill between</h3>

```python
plt.fill_between(x, y, color='skyblue', alpha=0.5)
plt.title("Fill Between")
plt.show()
```

<h3>Annotate points</h3>

```python
plt.plot(x, y)
plt.annotate("Peak", xy=(np.pi/2, 1), xytext=(2, 1.2),
             arrowprops=dict(facecolor='red', shrink=0.05))
plt.title("Annotation Example")
plt.show()
```

<h3>Save figure</h3>

```python
plt.plot(x, y)
plt.savefig("line_plot.png")
```

<h3>Gridlines</h3>

```python
plt.plot(x, y,color='green')
plt.grid(True)
plt.title("Gridlines Example")
plt.show()
```

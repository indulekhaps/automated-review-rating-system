<h3>5.Handling Missing Data</h3>

```python
df.isnull().sum()                     # Count nulls
```

```python
df.dropna(inplace=True)              # Drop rows with any nulls
df.head()
```

<h3>6.Removing Duplicates</h3>

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

<h3>7.Selecting Columns</h3>

```python
df['Score']                        # Single column (Series)
```

```python
df[['Score', 'Text']]              # Multiple columns (DataFrame)
```

<h3>8.Filtering Rows</h3>

```python
df[df['Score'] > 3]                          # score > 3
```

```python
df[df['ProductId'] == 'B001E4KFG0']                 
```

```python
df[(df['Score'] > 3) & (df['ProductId'] == 'B001E4KFG0')]  # Compound filter
```

<h3>9.Sorting Data</h3>

```python
df.sort_values(by='Score', ascending=True)
```

<h3>10.Creating New Columns</h3>

```python
df['PositiveReview'] = df['Score'] > 3
df.head()

```

```python
df['NewColumn'] = "Hello"   # every row will have "Hello"
df.head()
```

<h3>11.Dropping Columns or Rows</h3>

```python
df.drop('NewColumn', axis=1, inplace=True)        # Drop column
df.head()
```

<h3>12.Renaming Columns</h3>

```python
df.rename(columns={'Score': 'Ratings'}, inplace=True)
df.head()
```

<h3>13.Value Counts & Unique</h3>

```python
df['Time'].value_counts()
```

<h3>14.Grouping and Aggregation</h3>

```python
df.groupby('PositiveReview')['Ratings'].mean() #the average score for positive vs. non-positive reviews.
df.head()
```

<h3>String Operations</h3>

```python
df['ProfileName'].str.lower()
```

```python
df['ProfileName'].str.upper()
```

<h3>Mapping & Replacing Values</h3>

```python
# Example: Replace Score 1 with 0, and Score 5 with 5
df['HelpfulnessNumerator'] = df['HelpfulnessNumerator'].replace({1: 0, 5: 5})
df.head()
```

<h3>Save Cleaned Data</h3>
